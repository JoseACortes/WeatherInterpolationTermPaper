# Interpolation Method Code

## Legrange Method

```python
#function for individual legrange interpolation for each point
def legrange_interpolate(x, x_values, y_values):
    def _basis(j):
        p = [(x - x_values[m])/(x_values[j] - x_values[m]) for m in range(k) if m != j]
        return np.prod(p)
    k = len(x_values)
    return sum(_basis(j)*y_values[j] for j in range(k))
```

### Legrange Implementation

```python
daily_interpolation0 = []
_set = set0
_n = len(_set)
for x in range(0, _n, 1):
    daily_interpolation0.append(legrange_interpolate(x, [i for i in range(0, _n, 2)], [_set[i] for i in range(0, _n, 2)]))
```

## Neuton Method

```python
def NDD(x,y):
    n = len(x)
    A = np.zeros((n,n+1))
    A[:,0]= x[:]
    A[:,1]= y[:]
    for j in range(2,n+1):
        for i in range(j-1,n):
            A[i,j] = (A[i,j-1]-A[i-1,j-1]) / (A[i,0]-A[i-j+1,0])
    p = np.zeros(n)
    for k in range(0,n):
        p[k] = A[k,k+1]
    return p
def newton_polynomial(t,x,p):
    n = len(x)
    out = p[n-1]
    for i in range(n-2,-1,-1):
        out = out*(t-x[i]) + p[i]
    return out
```

```python
def NDD(x,y):
    n = len(x)
    A = np.zeros((n,n+1))
    A[:,0]= x[:]
    A[:,1]= y[:]
    for j in range(2,n+1):
        for i in range(j-1,n):
            A[i,j] = np.divide((A[i,j-1]-A[i-1,j-1]), (A[i,0]-A[i-j+1,0]))
    p = np.zeros(n)
    for k in range(0,n):
        p[k] = A[k,k+1]
    return p
def newton_polynomial(t,x,p):
    n = len(x)
    out = p[n-1]
    for i in range(n-2,-1,-1):
        out = out*(t-x[i]) + p[i]
    return out
```

### Neuton Implementation

```python
daily_interpolation1 = []
_set = set0
_n = len(_set)
_coeff = NDD([i for i in range(0, _n, 2)], [_set[i] for i in range(0, _n, 2)])
for x in range(0, _n, 1):
    daily_interpolation1.append(newton_polynomial(x, [i for i in range(0, _n, 2)], _coeff))
```

```python
hourly_interpolation1 = []
_set = set1
_n = len(_set)
_coeff = NDD([i for i in range(0, _n, 2)], [_set[i] for i in range(0, _n, 2)])
for x in range(0, _n, 1):
    hourly_interpolation1.append(newton_polynomial(x, [i for i in range(0, _n, 2)], _coeff))
```

```python
#daily two years
FREQUENCY = 24
START_DATE = '1-JAN-2018'
END_DATE = '31-DEC-2019'
API_KEY = key
LOCATION_LIST = ['76019']

datadaily = retrieve_hist_data(API_KEY,
                                LOCATION_LIST,
                                START_DATE,
                                END_DATE,
                                FREQUENCY,
                                location_label = False,
                                export_csv = True,
                                store_df = True)
```

```python
superdata = datadaily[0][['date_time', 'tempC']]
_ = [int(i) for i in datadaily[0]['tempC']]
superdata = superdata.reset_index(drop=True)
superdata['tempC'] = _
```

```python
import time
```

```python
_dat = superdata[:730]
nodes = []
avg_error = []
std_error = []
avg_error0 = []
std_error0 = []
legrange_time = []
neuton_time = []
for i in range(2, len(_dat), 2):
    #setup
    l_interpolation = []
    n_interpolation = []
    _set = _dat[:i]['tempC']
    _n = len(_set)
    _nodesx = [i for i in range(0, _n, 2)]
    _nodesy = [_set[i] for i in range(0, _n, 2)]
    #legrange
    l_startwatch = time.time()
    
    for x in range(0, _n, 1):
        l_interpolation.append(legrange_interpolate(x, _nodesx, _nodesy))
    
    l_stopwatch = time.time()
    legrange_time.append(l_stopwatch-l_startwatch)
    
    #neuton
    n_startwatch = time.time()
    
    _coeff = NDD(_nodesx, _nodesy)
    for x in range(0, _n, 1):
        n_interpolation.append(newton_polynomial(x, _nodesx, _coeff))
    
    n_stopwatch = time.time()
    neuton_time.append(n_stopwatch-n_startwatch)
    #interpolations are equal
    _interpolation = n_interpolation
    _interpolation0 = l_interpolation
    #counting nodes
    nodes.append(int(((i)/2)+1))
    #error
    #_error = abs((_set - np.array(_interpolation))/_set)
    _error = abs((_set - np.array(_interpolation)))
    _error0 = abs((_set - np.array(_interpolation0)))
    avg_error.append(np.mean(_error))
    std_error.append(np.std(_error))
    avg_error0.append(np.mean(_error0))
    std_error0.append(np.std(_error0))
_clean = {'node_count':nodes, 'l_time_s':legrange_time, 'n_time_s':neuton_time, 'avg_error':avg_error, 'std_error':std_error, 'avg_error0':avg_error0, 'std_error0':std_error0}
superlegrange = pd.DataFrame(_clean)
```
